<b>2.	Estimasi Kepekatan Parameterik  </b>

```python
# contoh parametric probability density estimation
# library yang digunakan 

from matplotlib import pyplot
from numpy.random import normal
from numpy import mean
from numpy import std
from scipy.stats import norm

# membangkitkan sample 
sample = normal(loc=50, scale=5, size=1000)

# menghitung parameters
sample_mean = mean(sample)
sample_std = std(sample)
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))

# mendefinisikan distribusi  
dist = norm(sample_mean, sample_std)
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]

# plotting histogram and pdf
pyplot.hist(sample, bins=10, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

<b>3.	Estimasi Kepekatan Nonparameterik </b>

```python
# Library yang diguanakan 
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity

#Ilustrasi kondisi data bimodal dengan dua ditribusi yang berbeda

	
# contoh kondisi bimodal data
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
# membangkitkan sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = hstack((sample1, sample2))
# plot histogram
pyplot.hist(sample, bins=50)
pyplot.show()

#Untuk melakukan fitting terhadap sample tersebut, kita gunakan fungsi kernal density estimation 

model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)

#Sebagai contoh yang digunakan adalah fungsi kepekatan kernal yang terdapa di dalam library scikit. Langkah berikutnya adalah menghitung peluang sample pada range tertentu 

# sample probabilities for a range of outcomes
values = asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)

#Setelah menentukan peluang sample tahap berikutya adalah melakukan plotting ke dalam kurva. 
# plot the histogram and pdf
pyplot.hist(sample, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()

```

2.	Pada estimasi kepekatan parametrik, cari distribusi lain selain distribusi normal dan lakukan visualiasi<br><br>
Exponential Distribution

```python
from matplotlib import pyplot
from numpy.random import exponential
from numpy import mean
from numpy import std
from scipy.stats import expon

# generate a sample from exponential distribution
sample_exp = exponential(scale=1, size=1000)

# calculate parameters
sample_mean_exp = mean(sample_exp)
sample_std_exp = std(sample_exp)
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean_exp, sample_std_exp))

# define the distribution
dist_exp = expon(scale=sample_mean_exp)
values_exp = [value for value in range(0, 10)]
probabilities_exp = [dist_exp.pdf(value) for value in values_exp]

# plot histogram and pdf
pyplot.hist(sample_exp, bins=10, density=True)
pyplot.plot(values_exp, probabilities_exp)
pyplot.title('Exponential Distribution - Parametric Probability Density Estimation')
pyplot.show()
```

3.	Buat histogram untuk contoh pembangkitan data dengan jumlah peak = 3 (multimodal) kemudian gunakan fungsi kernel density estimation dan lakukan visualisasi

```python
from matplotlib import pyplot
from numpy import hstack, asarray, exp
from numpy.random import normal
from sklearn.neighbors import KernelDensity

# generate a multimodal sample with three peaks
sample1_multi = normal(loc=20, scale=5, size=300)
sample2_multi = normal(loc=40, scale=5, size=700)
sample3_multi = normal(loc=60, scale=5, size=500)
sample_multi = hstack((sample1_multi, sample2_multi, sample3_multi))

# plot histogram
pyplot.hist(sample_multi, bins=50)
pyplot.title('Multimodal Data - Histogram with Three Peaks')
pyplot.show()

# apply kernel density estimation
model_multi = KernelDensity(bandwidth=2, kernel='gaussian')
sample_multi = sample_multi.reshape((len(sample_multi), 1))
model_multi.fit(sample_multi)

# sample probabilities for a range of outcomes
values_multi = asarray([value for value in range(1, 80)])
values_multi = values_multi.reshape((len(values_multi), 1))
probabilities_multi = model_multi.score_samples(values_multi)
probabilities_multi = exp(probabilities_multi)

# plot histogram and pdf
pyplot.hist(sample_multi, bins=50, density=True)
pyplot.plot(values_multi[:], probabilities_multi)
pyplot.title('Kernel Density Estimation - Multimodal Data')
pyplot.show()
```
