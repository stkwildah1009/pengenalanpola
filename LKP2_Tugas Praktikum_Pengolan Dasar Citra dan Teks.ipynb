**Pengolahan Dasar Citra:**

1. Membuka dan Menampilkan Citra

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
# Membaca citra
kopi = img.imread("kopi.jpg")

# Menampilkan citra
plt.imshow(kopi)
plt.axis('off')  # Menghilangkan sumbu koordinat
plt.show()
```

2. Mengubah Ukuran Citra

```python
# Mengubah ukuran citra
resize = (150, 100)
aresize = cv2.resize(kopi, resize)

# Menampilkan citra yang telah diubah ukurannya
plt.imshow(aresize)
plt.axis('off')  # Menghilangkan sumbu koordinat
plt.show()
```

3. Mengubah Citra menjadi Skala Abu-abu

```python
# Mengubah citra menjadi skala abu-abu
kopiabu = cv2.cvtColor(kopi, cv2.COLOR_RGB2GRAY)

# Menampilkan citra yang telah diubah ukurannya dan skala abu-abu
plt.imshow(kopiabu, cmap='gray')
plt.axis('off')  # Menghilangkan sumbu koordinat
plt.show()
```

4. Merotasi Citra

```python
# Memutar citra 180 derajat
#rotated_1 = kopi.rotate(180)
rotated_1 = cv2.rotate(kopi, cv2.ROTATE_180)

  
# This is Alternative Syntax To Rotate 
# The Image
#rotated_2 = kopi.transpose(Image.ROTATE_90)
rotated_2 = cv2.rotate(kopi, cv2.ROTATE_90_CLOCKWISE)

  
# This Will Rotate Image By 270 Degree
rotated_3 = cv2.rotate(kopi, cv2.ROTATE_90_COUNTERCLOCKWISE)


print("citra 180 derajat")
plt.imshow(rotated_1)
plt.axis('off')  
plt.show()
print("citra 90 derajat")
plt.imshow(rotated_2)
plt.axis('off')  
plt.show()
print("citra 270 derajat")
plt.imshow(rotated_3)
plt.axis('off')  # Menghilangkan sumbu koordinat
plt.show()
```

5. Image Histogram Equalization

```python
import numpy as np

plt.imshow(kopi)
plt.axis('off')  # Menghilangkan sumbu koordinat
plt.show()
hist,bins = np.histogram(kopi.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(kopi.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

6. Fourier Transform

```python
f = np.fft.fft2(kopi)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
plt.subplot(121),plt.imshow(kopi, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```
